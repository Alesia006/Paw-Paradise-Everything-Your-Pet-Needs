```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Të dhënat tona për Tiranën
te_dhenat = {
    'ora': [8, 9, 14, 15, 18, 3, 23, 12, 17, 7],
    'makinat': [500, 480, 200, 250, 650, 20, 50, 300, 600, 450],
    'trafiku': [2, 2, 1, 1, 2, 0, 0, 1, 2, 2] # 0=Low, 1=Medium, 2=High
}

df = pd.DataFrame(te_dhenat)

# Trajnimi i modelit
X = df[['ora', 'makinat']]
y = df['trafiku']
modeli = DecisionTreeClassifier()
modeli.fit(X, y)

# Këtu bëjmë provën: Shkruaj orën dhe sa makina po sheh semafori
ora_test = 11
makina_test = 20

parashikimi = modeli.predict([[ora_test, makina_test]])
etiketat = {0: "Low", 1: "Medium", 2: "High"}

print("---------------------------------------")
print(f"REZULTATI: Trafiku në Tiranë është: {etiketat[parashikimi[0]]}")
print("---------------------------------------")
```
